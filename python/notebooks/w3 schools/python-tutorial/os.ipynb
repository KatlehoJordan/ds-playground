## os

```python
print('The `os` module offers ways to interact with the operating system\n',
      '(OS) and shell')

import os

print('The following are the names imported by `os`:')

for key in dir(os):
    print(key)
```

### Sending commands to the shell

```python
print('To send a command to the shell, use `.system()` method')

import os

os.system('pwd')

print('This is handy and more robust than relying on jupyter for the ! syntax')

! pwd
```

### Simulating a crash

```python
print('the `_exit(1)` method will simulate a system crash, exiting the \n',
      'system prior to doing garbage cleanup in python. This may be useful \n',
      'for reproducing bugs that arise after a system crash.')

import os

# pyright: reportPrivateUsage=false

print(os._exit.__doc__)

print('Running the following will kill the Jupyter kernel, forcing you to\n',
      'restart it')

os._exit(1)
```
