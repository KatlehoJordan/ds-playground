## Math

```python
print('A few math functions are built in, including `min()` and `max()`')

test_set = (4, 1, 6, 2)

print(f'The test set is: {test_set}')
print(f'The minimum of the test_set is {min(test_set)}')
print(f'The maximum of the test_set is {max(test_set)}')

print('The `abs()` yields the absolute value')

print(abs(-4.2))

print('The `pow()` function raises a base to a power')

print(pow(4, 3))
```

### The Math Module

```python
print('The `math` module offers more functionality')

import math

print(math.sqrt(64))

print(math.ceil(4.2))
print(math.floor(4.2))
print(math.pi)
```
