## Seaborn Module

The official documentation [is here](https://seaborn.pydata.org/).

Seaborn uses Matplotlib to plot graphs. This means that for basic seaborn plots, it is sufficient to import seaborn only. However, to customize the plots in many standard ways, one needs to also import matplotlib in order to interact with that API.

```python
import numpy as np
import seaborn as sns # type: ignore
from configurations import logger

logger.warning(
    'It appears that seaborn is imported without a "stub file" so that its\n'
    'types cannot be identified by pylance. As a result I explicitly ignore\n'
    'types for lines with seaborn since they are unknown.'
)

my_array = np.array([*range(1, 6)])

sns.distplot(my_array) # type: ignore
```

Observe that the tutorial relies upon sns.distplot(), but this function is deprecated. Thus, I will modify function calls to use the updated sns API as much as I can.

Documentation regarding deprecating `sns.distplot()` is found [here](https://gist.github.com/mwaskom/de44147ed2974457ad6372750bbe5751).

Other potentially helpful links:
- [about `sns.displot()`](https://seaborn.pydata.org/generated/seaborn.displot.html)
- [about `sns.histplot()`](https://seaborn.pydata.org/generated/seaborn.histplot.html)
- [Stack overflow tip](https://stackoverflow.com/questions/65973224/updating-seaborn-distplot-code-to-version-0-11) to replace `distplot` with `histplot(kde=True, stat='density', linewidth=0)`
- [Blog post](https://medium.com/geekculture/seaborns-distplot-function-is-deprecated-what-to-use-instead-fa082b749bc6) indicating to replace `distplot` with `displot(kde=True)`

The best way I found to replicate the deprecated `distplot` is to also import `matplotlib.pyplot` in order to get the `subplots` method to create an `ax` object which can be shared by the `kdeplot` and the `histplot`.

```python
import numpy as np
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt
from configurations import logger

logger.warning(
    'It appears that seaborn is imported without a "stub file" so that its\n'
    'types cannot be identified by pylance. As a result I explicitly ignore\n'
    'types for lines with seaborn since they are unknown.'
)

logger.warning(
    'It appears that matplotlib subplots typing triggers the linter.\n'
    'As a result I explicitly ignore types in those lines.'
)

fig, ax = plt.subplots() # type: ignore

my_array = np.array([*range(1, 6)])

sns.kdeplot(my_array, ax=ax) # type: ignore
sns.histplot(  # type: ignore
    my_array,
    ax=ax,
    stat='density',
    binwidth=2,
    edgecolor='None',
    color='lightblue'
    )
```
