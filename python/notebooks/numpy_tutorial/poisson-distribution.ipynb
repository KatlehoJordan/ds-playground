## Poisson Distribution

The Wikipedia article does a better job describing this distribution: [here](https://en.wikipedia.org/wiki/Poisson_distribution).

It predicts the probability of event occurring in a given dimension (time, physical space, etc.) if there is a known and well-established mean rate of that event.

The mean is indicated with $\lambda$.

In `numpy.random.poisson()` the arguments are `lam` for lambda and `size` for the size of the array.

```python
import numpy as np
from configurations import printer

lambdas_to_cycle_through = [1, 10, 100]
for lam in lambdas_to_cycle_through:
    my_array = np.random.poisson(lam=lam, size = 10)
    printer(
        'These are 10 examples of the results from a Poisson distribution\n'
        'with lambda = %s:\n%s\n'
        'They have mean: %s',
        lam,
        my_array,
        np.mean(my_array)
        )
```

### Visualization of Poisson distribution

```python
import numpy as np
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt
from configurations import logger

plt.style.use('my_mpl_stylesheet.mpltstyle')

logger.warning(
    'Seaborn does not seem to export a stub to help linters with type hints\n'
    'so I am explicitly disabling type hints on those lines.'
)

logger.warning(
    'Matplotlib.pyplot does not seem to have types set up for linters.\n'
    'So I am explicitly disabling type hints on those lines.'
)

def plot_poisson(
        lam: int = 1,
        size: int = 1_000,
        bw_adjust: float = 1
        ) -> None:
    my_poisson = np.random.poisson(lam=lam, size=size)

    sns.histplot( # type: ignore
        my_poisson,
        ax=plt.subplots()[1], # type: ignore
        stat='probability',
        kde=True,
        kde_kws={'bw_adjust' : bw_adjust},
        edgecolor='None'
        )

    plt.title( # type: ignore
        label=f'{size:,} samples from a Poisson distribution with\n'
        fr'$\lambda$={lam:,}',
        family='serif'
        )

    plt.xlabel(xlabel='Value of sample', family='sans-serif') #type: ignore
    
plot_poisson(size=1_000)

logger.warning(
    'The kernel density plot is misleading; it shows small dips between the\n'
    'discrete values, but the probability of values between those values is\n'
    '0. Thus, although there are really dips in those areas, the dips should\n'
    'go to 0. A more accurate depiction can be made by modifying the\n'
    'bandwidth of the kernel-density estimate (KDE).'
)

plot_poisson(size=1_000, bw_adjust=0.6)
```

Observe that a Poisson distribution can be approximated by a normal distribution where:

$\mu=\lambda$

and

$\sigma=\sqrt{\lambda}$

However, the normal approximation is more accurate as $\lambda$ increases.

```python
import numpy as np
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt
from configurations import logger

plt.style.use('my_mpl_stylesheet.mpltstyle')

logger.warning(
    'Seaborn does not seem to export a stub to help linters with type hints\n'
    'so I am explicitly disabling type hints on those lines.'
)

logger.warning(
    'Matplotlib.pylot does not seem to have types set up for linters.\n'
    'So I am explicitly disabling type hints on those lines.'
)

def choose_my_color(integer: int = 0) -> str:
    return (plt.rcParams["axes.prop_cycle"] # type: ignore
            .by_key()["color"][integer])

def plot_poisson_and_normal(
        lam: int = 1,
        size: int = 1_000,
        bw_adjust: float = 1
        ) -> None:
    my_poisson = np.random.poisson(lam=lam, size=size)

    ax = plt.subplots()[1] # type: ignore

    sns.histplot( # type: ignore
        my_poisson,
        ax=ax,
        stat='probability',
        kde=True,
        kde_kws={'bw_adjust' : bw_adjust},
        edgecolor='None',
        color=choose_my_color(0),
        label='Poisson'
        )

    loc = lam
    scale = np.sqrt(lam)

    my_normal = np.random.normal(loc=loc, scale=scale, size=size)

    sns.histplot( # type: ignore
        my_normal,
        ax=ax,
        stat='probability',
        kde=True,
        kde_kws={'bw_adjust' : bw_adjust},
        edgecolor='None',
        color=choose_my_color(5),
        label='Normal'
        )
    
    ax.legend() # type: ignore

    plt.title( # type: ignore
        label=f'{size:,} samples from:\n'
        'a Poisson distribution with '
        fr'$\lambda$={lam:,}'
        '\nand\n'
        'a normal distribution with '
        fr'$\mu$={loc:,.0f} and $\sigma\approx${scale:.1f}',
        family='serif'
        )

    plt.xlabel(xlabel='Value of sample', family='sans-serif') #type: ignore

lambdas_bandwidths_to_cycle_through = {
    1 : 0.7,
    10 : 0.5,
    100 : 0.3,
    1_000 : 0.1
    }

for lam in lambdas_bandwidths_to_cycle_through:
    plot_poisson_and_normal(
        lam=lam,
        bw_adjust=lambdas_bandwidths_to_cycle_through[lam]
        )
```

Because a binomial distribution can be approximated by a normal distribution where:

$\mu = n * p$

and

$\sigma = \sqrt{n * p * (1 - p)}$

and a Poisson distribution can be approximated by a normal distribution where:

$\mu = \lambda$

and

$\sigma = \sqrt{\lambda}$

Thus:

$\lambda = n * p$

and

$\lambda = n * p * (1 - p)$

Thus:

$n * p = n * p * (1 - p)$

Which simplifies to:

$1 = (1 - p)$

or:

$p = 0$

This means that the binomial distribution approximates the Poisson distribution when $p$ is very near 0.

To show this, $n$ has to be sufficiently large. From the above equalities, one can see that:

$n=\lambda/p$

```python
import numpy as np
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt
from configurations import logger

plt.style.use('my_mpl_stylesheet.mpltstyle')

logger.warning(
    'Seaborn does not seem to export a stub to help linters with type hints\n'
    'so I am explicitly disabling type hints on those lines.'
)

logger.warning(
    'Matplotlib.pylot does not seem to have types set up for linters.\n'
    'So I am explicitly disabling type hints on those lines.'
)

def choose_my_color(integer: int = 0) -> str:
    return (plt.rcParams["axes.prop_cycle"] # type: ignore
            .by_key()["color"][integer])

def plot_poisson_and_binomial(
        lam: int = 1,
        size: int = 1_000,
        p: float = 0.5,
        bw_adjust: float = 1
        ) -> None:
    my_poisson = np.random.poisson(lam=lam, size=size)

    ax = plt.subplots()[1] #type: ignore

    sns.histplot( # type: ignore
        my_poisson,
        ax=ax,
        stat='probability',
        kde=True,
        kde_kws={'bw_adjust' : bw_adjust},
        edgecolor='None',
        color=choose_my_color(0),
        label='Poisson'
        )
    
    n = np.ceil(lam / p)
    
    my_binomial = np.random.binomial(n=n, p=p, size=size)

    sns.histplot( # type: ignore
        my_binomial,
        ax=ax,
        stat='probability',
        kde=True,
        kde_kws={'bw_adjust' : bw_adjust},
        edgecolor='None',
        color=choose_my_color(5),
        label='Binomial'
        )
    
    ax.legend() # type: ignore

    plt.title( # type: ignore
        label=f'{size:,} samples from:\n'
        'a Poisson distribution with '
        fr'$\lambda$={lam:,}'
        '\nand\n'
        'a binomial distribution with '
        fr'$n$={n:,.0f} and $p$={p}',
        family='serif'
        )

    plt.xlabel(xlabel='Value of sample', family='sans-serif') #type: ignore

lambdas_bandwidths_to_cycle_through = {
    1 : 0.7,
    10 : 0.5,
    100 : 0.3,
    1000 : 0.3
    }

probabilities_to_cycle_through = [0.99,
                                  0.5,
                                  0.01]

for lam in lambdas_bandwidths_to_cycle_through:
    for probability in probabilities_to_cycle_through:
        plot_poisson_and_binomial(
            lam=lam,
            p=probability,
            bw_adjust=lambdas_bandwidths_to_cycle_through[lam]
            )
```
