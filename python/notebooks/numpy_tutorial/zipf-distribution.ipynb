## Zipf Distribution

This is a discrete distribution, and describes sets where when items are sorted in order of decreasing frequency, the $nth$ entry is inversely proportional to $n$. For example, for natural language:

$word \space frequency \propto 1/word \space rank$

See the [Wikipedia article here](https://en.wikipedia.org/wiki/Zipf%27s_law).

It takes on one parameter, `a` which must be a float value greater than 1. As `a` increases, so the distribution becomes more and more stacked on the first value (the tail gets 'sucked in' towards the first items in the distribution.)

```python
import numpy as np
from configurations import printer

as_to_cycle_through = [1.1, 1.2, 10]
for a in as_to_cycle_through:
    size = 10
    my_array = np.random.zipf(a=a, size=size) # type: ignore
    printer(
        'This is an array from a Zipf distribution of size %s with\n'
        'shape parameter a %s\n:%s\n',
        size,
        a,
        my_array
    )
```

### Visualization of the Zipf distribution

The distribution is both discrete and monotonically decreasing, so the standard kernel-density estimate would be misleading.

```python
import numpy as np
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt
from configurations import logger

plt.style.use('my_mpl_stylesheet.mpltstyle')

logger.warning(
    'Seaborn does not seem to export a stub to help linters with type hints\n'
    'so I am explicitly disabling type hints on those lines.'
)

logger.warning(
    'Matplotlib.pyplot does not seem to have types set up for linters.\n'
    'So I am explicitly disabling type hints on those lines.'
)

def plot_zipf(
        a: float = 2,
        size: int = 100
    ) -> None:

    my_zipf = np.random.zipf(a=a, size=size) # type: ignore
    
    ax = plt.subplots()[1] # type: ignore
    
    sns.histplot( # type: ignore
        my_zipf,
        ax=ax,
        edgecolor='None'
        )
    
    plt.title( # type: ignore
        label=f'{size:,} samples from a Zipf distribution with\n'
        fr'$\alpha=${a:,.2f}'
        f'\nThe maximum value in the sample is {np.max(my_zipf)}',
        family='serif'
        )
    
    plt.xlabel(xlabel='Value of sample', family='sans-serif') # type: ignore

    ax.set_xlim(0, 10)

as_to_cycle_through = [2, 3, 4, 5]
for a in as_to_cycle_through:
    plot_zipf(a=a)
```
