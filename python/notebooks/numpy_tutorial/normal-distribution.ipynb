## Normal Distribution

Use `random.normal()` to get a normal distribution. Specify these three parameters:
- `loc` = mean
- `scale` = the standard deviation
- `size` = the shape of the returned array

```python
import numpy as np
from configurations import printer

my_2_by_3_of_z = np.random.normal(loc=0, scale=1, size = (2, 3))

printer(my_2_by_3_of_z)

my_2_by_3_of_normal = np.random.normal(loc=1, scale=2, size=(2, 3))

printer(my_2_by_3_of_normal)
```

### Visualization of Normal distribution

```python
import numpy as np
import seaborn as sns # type: ignore
import matplotlib as mpl
import matplotlib.pyplot as plt
from configurations import logger

titles_and_labels_font = 'serif'
ticks_font = 'monospace'
ticks_fontsize = 6
mpl.rc('font', family=titles_and_labels_font) # type: ignore

logger.warning(
    'Seaborn does not seem to export a stub to help linters with type hints\n'
    'so I am explicitly disabling type hints on those lines.'
)

logger.warning(
    'Matplotlib does not seem to have types set up for linters.\n'
    'So I am explicitly disabling type hints on those lines.'
)

loc=0
scale=1
size=1_000
my_normal = np.random.normal(loc=loc, scale=scale, size=size)

ax1 = plt.subplots()[1] # type: ignore
ax2 = plt.twinx()
sns.histplot( # type: ignore
    my_normal,
    ax=ax1,
    stat='probability',
    color='lightblue',
    edgecolor='None',
    label='Normal'
    )
sns.kdeplot( # type: ignore
    my_normal,
    ax=ax2,
    color='blue',
    label='Normal KDE'
    )

plt.title( # type: ignore
    f'{size:,} samples from a normal distribution with \n'
    f'mean={loc} and standard deviation={scale}'
    )

ax1.set_xlabel('Value of sample') # type: ignore
ax1_xticks = ax1.get_xticks()
ax1.set_xticks(ax1_xticks)
formatted_ax1_xticks = ["{:.0f}".format(val) for val in ax1_xticks]
ax1.set_xticklabels(
    formatted_ax1_xticks,
    fontproperties=ticks_font,
    fontsize=ticks_fontsize
    )

ax1.set_ylabel('Probability of value') # type: ignore
ax1_yticks = ax1.get_yticks()
ax1.set_yticks(ax1_yticks)
formatted_ax1_yticks = ["{:.2f}".format(val) for val in ax1_yticks]
ax1.set_yticklabels(
    formatted_ax1_yticks,
    fontproperties=ticks_font,
    fontsize=ticks_fontsize
    )
ax1.legend(loc='upper left') # type: ignore

ax2.set_ylabel( # type: ignore
    'Density (Probability per value)',
    rotation=270,
    labelpad=15
    )
ax2_yticks = ax2.get_yticks()
ax2.set_yticks(ax2_yticks)
formatted_ax2_yticks = ["{:.2f}".format(val) for val in ax2_yticks]
ax2.set_yticklabels(
    formatted_ax2_yticks,
    fontproperties=ticks_font,
    fontsize=ticks_fontsize
    )
ax2.legend(loc='upper right') # type: ignore

sns.despine(right=False) # type: ignore
```

Observe how using more samples makes the plot look more like the underlying distribution.

```python
import numpy as np
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt
from configurations import logger

plt.style.use('my_mpl_stylesheet.mpltstyle')

logger.warning(
    'Seaborn does not seem to export a stub to help linters with type hints\n'
    'so I am explicitly disabling type hints on those lines.'
)

logger.warning(
    'Matplotlib subplots does not seem to have types set up for linters.\n'
    'So I am explicitly disabling type hints on those lines.'
)

def plot_normal(size: int = 10) -> None:
    my_normal = np.random.normal(size=size)

    ax = plt.subplots()[1] # type: ignore
    sns.kdeplot( # type: ignore
        my_normal,
        ax=ax,
        label='Normal'
        )
    sns.histplot( # type: ignore
        my_normal,
        ax=ax,
        color='lightblue',
        edgecolor='None',
        stat='density',
        label='Normal'
        ).set(title=f'Normal distribution with {integer:,} samples')
    sns.despine() # type: ignore
    plt.legend() # type: ignore
    
for integer in [10, 100, 1_000, 10_000]:
    plot_normal(integer)
```
